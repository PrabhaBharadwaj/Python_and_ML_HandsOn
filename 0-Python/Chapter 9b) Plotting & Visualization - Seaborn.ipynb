# Figure Aesthetic (set_style)

```python
def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5): 
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
# matplotlib way of creation

sinplot()
plt.show()
```

```python
# To change the same plot to Seaborn defaults, use the set() function 
# The default theme of the plot will be darkgrid 

sb.set()
sinplot()
plt.show()
```

```python
# Change it to white grid:

sb.set_style("whitegrid")
sinplot()
plt.show()
```

```python
# Removing Axes Spines

sb.set_style("white")
sinplot()
sb.despine() # we can remove the top and right axis spines using the despine() function.
plt.show()
```

```python
# To customize the Seaborn styles, you can pass a dictionary of parameters to the set_style() function.

sb.set_style("darkgrid", {'axes.axisbelow': False})  # makes grid lines prominent, self observation
sinplot()
sb.despine()
plt.show()
```

```python
sb.set_style("darkgrid", {'axes.axisbelow': True}) # now grid lines are below the waves.
sinplot()
sb.despine()
plt.show()
```

# Color Palette

```python
# A palette means a flat surface on which a painter arranges and mixes paints.

seaborn.color_palette(palette = None, n_colors = None, desat = None)
```

```python
# A) Qualitative Color Palettes or Categorical palettes
# categorical palettes are best suitable to plot the categorical data.

current_palette = sb.color_palette()
sb.palplot(current_palette) #This function plots the color palette as horizontal array. 
plt.show()
```

```python
# B) Sequential Color Palettes

Sequential plots are suitable to express the distribution of data ranging from 
relative lower values to higher values within a range.
```

```python
sb.palplot(sb.color_palette("Greens"))
plt.show()
```

```python
# C) Diverging Color Palette

Each color represents variation in the value ranging from a common point in either direction.

Assume plotting the data ranging from -1 to 1. The values from -1 to 0 takes one color and 0 to +1 takes another color.

By default, the values are centered from zero.
```

```python
sb.palplot(sb.color_palette("BrBG", 7))
plt.show()
```

```python
# D) Setting the Default Color Palette

def sinplot(flip = 1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5):
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sb.set_style("white")
sb.set_palette("husl")
sinplot()
plt.show()
```
