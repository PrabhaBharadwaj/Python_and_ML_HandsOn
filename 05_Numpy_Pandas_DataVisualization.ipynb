```python
#Split and individual display
import numpy as np

new_series2 = pd.Series(['India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture'])
print(new_series2.values)

print("\n split the individual strings wherever ‘_’ comes and create a list out of it.:")
print(new_series2.str.split('_').values)

print("\n Access the individual element of a list:")
x = new_series2.str.split('_').values

for item in x:
    print(item)

    if str(item) == 'nan':
        print('nan')
    else: 
        for i in range(len((item))):
            print(item[i])
```

```python
# Multiline plot
# multiple plots(functions) can be created in one.
# Method 1: using multiple calls for plt.plot() method
x = [0,1,2,3,4]

plt.plot(x, [elem for elem in x])
plt.plot(x, [elem**2 for elem in x])
plt.plot(x, [elem**3 for elem in x])
plt.show()
```

```python
# method 2: plotting multiple curves by using single method call

plt.plot(x,[elem for elem in x],
         x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.show()
```

```python
# adding a Grid
plt.plot(x,[elem for elem in x],
         x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
plt.show()
```

```python
# Limiting the Axes
# specifying the range for x-axis ang y-axis values

x = [0,1,2,3,4]
plt.plot(x,[elem for elem in x],
         x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)
# Method 1
plt.axis([-1,3,-1,10])  
plt.show()
```

```python
# Limiting the Axes
# Method 2
x = [0,1,2,3,4]
plt.plot(x,[elem for elem in x],
         x, [elem**2 for elem in x],
        x, [elem**3 for elem in x])
plt.grid(True)

#method 2
plt.xlim(0,4)
plt.ylim(0,20)
plt.show()
```

## Showing percentage in the pie chart

```python
import matplotlib.pyplot as plt

#autopct will give each blocks values in %, %1.2f%%, here .2 means 2 decimal

names = ['aa', 'bb', 'cc', 'dd']

plt.pie([30, 25, 40, 35],
        labels=names,
        autopct='%1.2f%%'
)
plt.show()
```

```python
# using explode(): to separate a part of a pie chart

import matplotlib.pyplot as plt

names = ['aa', 'bb', 'cc', 'dd']

fig1, ax1 = plt.subplots()
ax1.pie([30, 25, 40, 35],
        explode=(0, 0.1, 0, 0), # it would separate 'bb'
        labels=names,
        autopct='%1.1f%%'
)

ax1.axis('equal')
plt.show()

```

```python
# using explode(): to separate a part of a pie chart far away

import matplotlib.pyplot as plt

names = ['aa', 'bb', 'cc', 'dd']

fig1, ax1 = plt.subplots()
ax1.pie([30, 25, 40, 35],
        explode=(0,2, 0, 0), # it would separate 'bb'
        labels=names,
        autopct='%1.1f%%'
)

ax1.axis('equal')
plt.show()

```

# SEABORN Heat map for confusion Matrix


```python
#Data import and build dataset
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

data = pd.read_csv(r"..\Dataset\pacific.csv")
                   
# data  wrangling
pred_columns = data[:]
pred_columns.drop(['Date', 'Time', 'ID','Name','Event','Status','Latitude','Longitude'],
                 inplace=True,
                 axis=1)
# store the relevant columns
relevant_columns = pred_columns.columns

#train test split
train, test = train_test_split(data, test_size=0.3)
train_X = train[relevant_columns]
train_Y = train['Status']

#testing data
test_X = test[relevant_columns]
test_Y = test['Status']
```

```python
#Apply Naivey Bayes Theorem
gnb_model = GaussianNB()
gnb_model.fit(train_X, train_Y)

#testing
predicted_op = gnb_model.predict(test_X)
print(accuracy_score(test_Y, predicted_op))
```

```python
a= confusion_matrix(test_Y, predicted_op)
print("\n Confusion Mtarix:\n" , a)
print("\n Transpose of Confusion Mtarix:\n" , a.T)
```

```python
#Plot seaborn heatmap for Confusion matrix

#labels = gnb_model.predict(test_X)  This is not used here
mat = confusion_matrix(test_Y, predicted_op)
sns.heatmap(mat.T,
            fmt='d',
            square=True,
            cbar=True,
            annot=True,
            xticklabels=relevant_columns,
            yticklabels=relevant_columns)
```
