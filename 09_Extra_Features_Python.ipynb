###  Imputation of missing values
    - For various reasons, many real world datasets contain missing values, often encoded as blanks, NaNs or other placeholders
    - A basic strategy to use incomplete datasets is to discard entire rows and/or columns containing missing values. 
    - However, this comes at the price of losing data which may be valuable (even though incomplete)
    
    SimpleImputer class	
            The SimpleImputer class provides basic strategies for imputing missing values
            Missing values can be imputed with a provided constant value, or using the statistics (mean, median or most frequent) of each column in which the missing values are located.



```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

##here used fit and transform separately
imp.fit([[1, 2], [np.nan, 3], [7, 6]])       
          
```

```python
##here transform applied on above foit value
X = [[np.nan, 2], [6, np.nan], [7, 6]]
print(X)
print("***")
print(imp.transform(X)) 
```

```python
# Replace the missing values by the most frequent value in each column
#for this we use SimpleImputer

df = pd.DataFrame([["a", "x"],
                   [np.nan, "y"],
                   ["a", np.nan],
                 ["b", "y"]], dtype="category")
df
```

# iNDIAN Rangoli

You are given an integer, N. Your task is to print an alphabet rangoli of size . (Rangoli is a form of Indian folk art based on creation of patterns.)

Different sizes of alphabet rangoli are shown below:
    
Input (stdin)
5

Your Output (stdout)

        --------e--------
        ------e-d-e------
        ----e-d-c-d-e----
        --e-d-c-b-c-d-e--
        e-d-c-b-a-b-c-d-e
        --e-d-c-b-c-d-e--
        ----e-d-c-d-e----
        ------e-d-e------
        --------e--------

```python
import string
def print_rangoli(size):
    order = size - 1
    for i in range(order, 0, -1):
        row = ['-'] * (2 * size - 1)
       
        for j in range(size - i):
            row[order - j] = row[order + j] = string.ascii_lowercase[j + i]
           
        print('-'.join(row))

    for i in range(0, size):
        row = ['-'] * (2 * size - 1)
        for j in range(0, size - i):
            row[order - j] = row[order + j] = string.ascii_lowercase[j + i]
        print('-'.join(row))

print_rangoli(5)
```

## pandas.melt  (UNPIVOT)  AND PIVOT
    - pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None)
    - id_vars  -->Column(s) to use as identifier variables. [ what we do not wish to melt]
    - value_vars  --> Column(s) to unpivot. If not specified, uses all columns that are not set as id_vars [what we want to melt]
    - var_name  ---> Name to use for the ‘variable’ column. If None it uses frame.columns.name or ‘variable’.
    - value_name --> Name to use for the ‘value’ column.
    - col_level  ---> If columns are a MultiIndex then use this level to melt.
    
    
    ## After melt it gives one or more as is col which mentioned in id_vars. and new col "variable" and "value"
    ### Variable will have "value_vars" column name as value and  "value" will have all values 

```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                 'B': {0: 1, 1: 3, 2: 5},
                  'C': {0: 2, 1: 4, 2: 6}})
df
```

```python
pd.melt(df, id_vars=['A'], value_vars=['B'])
```

```python
melted = pd.melt(df, id_vars=['A'], value_vars=['B', 'C'])
melted
```

```python
#The names of ‘variable’ and ‘value’ columns can be customized:
pd.melt(df, id_vars=['A'], value_vars=['B'],
        var_name='myVarname', value_name='myValname')
```

###  STACKING and UNSTACKING DATAFRAME
- deal with multi label indexes
- level: position of unstacked index
- swaplevel: change inner and outer level index position

```python
dic = {"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,4,72,65]}
df = pd.DataFrame(dic)
df
```

```python
df1 = df.set_index(["treatment","gender"]) #This called as stacking mostly
df1
```

```python
# lets unstack it
# level determines indexes  level=0 1st index in Hirarchy index, level=1 is 2nd index in  Hirarchy index
df1.unstack(level=0)
```

```python
df1.unstack(level=1)
```

```python
# change inner and outer level index position
df2 = df1.swaplevel(0,1)
df2
```

### SORTED IN Python

- Sort the Dictionary using Sorted function

```python
# nAME AND AGE
Sample = { 'Prabha' : 32, 'Yogi' : 33, 'Atharv': 2, 'Pranamya': 6, 'Pallu': 5 }
```

```python
sorted(Sample, key = Sample.get)
# Here key=Sample.get, sort the Dict based on the value in that dict, Deafault asc ord)
```

```python
sorted(Sample, key = Sample.get, reverse = True)  
# Here key=Sample.get, sort the Dict based on the value in that dict, here reverse =True so sort in desc)
```

```python
name_list =  sorted(Sample, key=Sample.get, reverse=True)

for name in name_list:
    # Here :>25  provides 25 letter space before item. Here :>5  provides 5 letter space before item count
    print(f"{name:>25}{Sample[name]:>5}")
```

```python
#Find top 2 elder

print("Aged 2 people")
for name in name_list[:2]:  # Starting 2  its in ascending order
    print(f"{name:20} Age : {Sample[name]:.2f}")

print("\n Minor 2 people")
# Find top 2 little one
for name in name_list[-2:]:  # Starting 10  its in ascending order
    print(f"  {name:20} Age : {Sample[name]:.2f}")
```

# f-Strings: A New and Improved Way to Format Strings in Python

- Also called “formatted string literals,” f-strings are string literals that have an f at the beginning and curly braces containing expressions that will be replaced with their values. The expressions are evaluated at runtime and then formatted using the __format__ protocol.

```python
name = "Eric"
age = 74
f"Hello, {name}. You are {age}."
```

```python
print(f"Hello, {name}. You are {age}.")
```

```python
print(f"Hello, {name:>10}. You are {age}.")
 # Here :>10  provides 10 letter space before item
```

```python
print(f"Hello, {name:<10}. You are {age}.")
 # Here :<10  provides 10 letter space after item
```
