# Hierarchial Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import silhouette_score
```

# Load the Dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset.head()
```

```python
dataset.isna().sum()
```

```python
# we need to see the clustering between Annual Income and Spending, hence we will load only those 2 columns. 

X = dataset.iloc[: , 3: ].values
```

# Draw the Dendogram

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X , method='ward'))
plt.title('Dendrogram')
plt.xlabel('customer')
plt.ylabel('Euclidean')
plt.show()

# we can take either as 5 clusters (3 orange line, 2 blue line) or 3 clusters ( 3 blue line). 
```

# Build the Model

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=5 , affinity='euclidean' , linkage='ward')
y_hc = hc.fit_predict(X)

# here we can change the affinitiy and linkage for hyper parameter tuning. 
# affinity is the algorithm used to find the distance between the data points. 
# linkage is the algorithm used to find the distance between the clusters. 
# ward works only with euclidean
```

```python
y_hc
```

```python
membership = hc.labels_
membership
```

```python
silhouette_score(X,membership)

# The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. 
# Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster 
# is more similar.
```

```python
plt.scatter(X[y_hc==0 , 0] , X[y_hc==0 , 1] , s=100 , c='red' , label='cluster 1')
plt.scatter(X[y_hc==1 , 0] , X[y_hc==1 , 1] , s=100 , c='pink' , label='cluster 2')
plt.scatter(X[y_hc==2 , 0] , X[y_hc==2 , 1] , s=100 , c='yellow' , label='cluster 3')
plt.scatter(X[y_hc==3 , 0] , X[y_hc==3 , 1] , s=100 , c='violet' , label='cluster 4')
plt.scatter(X[y_hc==4 , 0] , X[y_hc==4 , 1] , s=100 , c='blue' , label='cluster 5')
plt.xlabel('Salary')
plt.ylabel('Amount spend')
plt.title('Hierarchical Clustering')
plt.legend()
plt.show()
```

```python
# Let's try cluster = 3
```

```python
hc = AgglomerativeClustering(n_clusters=3 , affinity='euclidean' , linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
silhouette_score(X,membership)
```

```python
plt.scatter(X[y_hc==0 , 0] , X[y_hc==0 , 1] , s=100 , c='red' , label='cluster 1')
plt.scatter(X[y_hc==1 , 0] , X[y_hc==1 , 1] , s=100 , c='pink' , label='cluster 2')
plt.scatter(X[y_hc==2 , 0] , X[y_hc==2 , 1] , s=100 , c='yellow' , label='cluster 3')

plt.xlabel('Salary')
plt.ylabel('Amount spend')
plt.title('Hierarchical Clustering')
plt.legend()
plt.show()

# red one looks as very big cluster so 5 cluster only looks meaningful. 
```
