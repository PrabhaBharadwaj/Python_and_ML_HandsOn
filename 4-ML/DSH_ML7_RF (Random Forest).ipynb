# Load the data

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
diabetes.feature_names
```

```python
print(diabetes.DESCR)
```

```python
X = diabetes.data
y = diabetes.target

X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Random Forest as a Regressor 

```python
regressor = RandomForestRegressor(n_estimators=100, random_state = 42)

regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
```

```python
# y_test v/s y_pred plot

plt.figure(figsize=(16,4))
plt.plot(y_pred,  label='y_pred')
plt.plot(y_test, label='y_test')
plt.xlabel('X_test', fontsize=14)
plt.ylabel('Value of y(pred , test)', fontsize=14)
plt.title('Comparing predicted values and true values')
plt.legend(title='Parameter where:')
plt.show()
```

```python
def regression_metrics_udf(y_test, y_predict):
    
    print(' r2_score: ', r2_score(y_test, y_predict))

    print('\n\n Mean Absolute Error (MAE): ', mean_absolute_error(y_test, y_predict))

    print('\n\n Median Absolute Error (MdAE): ', median_absolute_error(y_test, y_predict))

    print('\n\n Mean Squared Error (MSE): ', mean_squared_error(y_test, y_predict))
    
    print('\n\n Root Mean Squared Error (RMSE): ', np.sqrt(mean_squared_error(y_test, y_predict)))

    print('\n\n Explained Variance Score (EVS): ', explained_variance_score(y_test, y_predict))

    print('\n\n Max Error (ME): ', max_error(y_test, y_predict))
```
